```python
# 👇🏻 YOUR CODE 👇🏻:
def get_yearly_revenue(monthly_revenue):
    return 12*monthly_revenue

def get_yearly_expenses(monthly_expenses):
    return 12*monthly_expenses

def get_tax_amount(profit):
    if profit > 100000:
        return profit*0.25
    else:
        return profit*0.15
    
def apply_tax_credits(tax_amount,tax_credits):
    amount_to_discount = tax_amount * tax_credits
    return amount_to_discount
# /YOUR CODE

# BLUEPRINT | DONT EDIT

monthly_revenue = 5500000
monthly_expenses = 2700000
tax_credits = 0.01

yearly_revenue = get_yearly_revenue(monthly_revenue)
yearly_expenses = get_yearly_expenses(monthly_expenses)

profit = yearly_revenue - yearly_expenses

tax_amount = get_tax_amount(profit)

final_tax_amount = tax_amount - apply_tax_credits(tax_amount, tax_credits)

print(f"Your tax bill is: ${final_tax_amount}")

# /BLUEPRINT
```

```python
def plus(num_1=0, num_2=0):
    return num_1 + num_2

def minus(num_1=0, num_2=0):
    return num_1 - num_2

def multi(num_1=0, num_2=0):
    return num_1 * num_2

def divi(num_1=0, num_2=1):
    return num_1 / num_2

while True:
    user_input_1 = input("Choose a number (or type 'exit' to finish): ")
    if user_input_1.lower() == 'exit':  # 'exit'을 입력하면 종료
        print("프로그램을 종료합니다.")
        break

    user_input_2 = input("Choose another one (or type 'exit' to finish): ")
    if user_input_2.lower() == 'exit':  # 'exit'을 입력하면 종료
        print("프로그램을 종료합니다.")
        break

    try:
        num_1 = int(user_input_1)  # 입력값을 정수로 변환
        num_2 = int(user_input_2)  # 입력값을 정수로 변환
    except ValueError:
        print("유효한 숫자를 입력하거나 'exit'을 입력하여 종료하세요.")
        continue

    print("Choose an operation:")
    print("     Options are: + , - , * , or /.")
    print("     Write 'exit' to finish.")
    operator = input()

    if operator.lower() == 'exit':
        print("프로그램을 종료합니다.")
        break
    elif operator == '+':
        result = plus(num_1, num_2)
        print(f'Result: {result}')
    elif operator == '-':
        result = minus(num_1, num_2)
        print(f'Result: {result}')
    elif operator == '*':
        result = multi(num_1, num_2)
        print(f'Result: {result}')
    elif operator == '/':
        if num_2 == 0:
            print("0으로 나눌 수 없습니다. 다시 시도해주세요.")
        else:
            result = divi(num_1, num_2)
            print(f'Result: {result}')
    else:
        print("유효한 연산자를 입력하세요.")
```
