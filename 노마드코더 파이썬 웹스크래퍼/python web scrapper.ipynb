```python
from requests import get

websites=[
"google.com",
"https://httpstat.us/502",
"https://httpstat.us/404",
"https://httpstat.us/300",
"https://httpstat.us/200",
"https://httpstat.us/101"
]
#https://httpstat.us/xxx 는 http코드를 생성해주는 사이트입니다

results={}

for website in websites:
    if not website.startswith("https://"):
        website = f"https://{website}"
    code = get(website).status_code
    if code >= 500:
        results[website] = "5xx / server error"
    elif code >= 400:
        results[website] = "4xx / client error"
    elif code >= 300:
        results[website] = "3xx / redirection "
    elif code >= 200:
        results[website] = "2xx / successful"
    elif code >= 100:
        results[website] = "1xx / informational response"

print(results)
```

```python
class Player:
    def __init__(self, name, team):
        self.name = name
        self.xp = 1500
        self.team = team

    def introduce(self):
        print(f"Hello! I'm {self.name} and I play for {self.team}")

class Team:
    def __init__(self, team_name):
        self.team_name = team_name
        self.players = []

    def show_players(self):
        for player in self.players:
            player.introduce()

    def add_player(self, name):
        new_player = Player(name, self.team_name)
        self.players.append(new_player)

    def remove_player(self, name):
        # 플레이어 이름으로 리스트에서 찾고 제거
        for player in self.players:
            if player.name == name:
                self.players.remove(player)
                print(f"{name} has been removed from {self.team_name}")
                return
        # 해당 이름의 플레이어가 없을 경우 메시지 출력
        print(f"Player with name '{name}' not found in team '{self.team_name}'.")

    def show_XP(self):
        total_xp = sum(player.xp for player in self.players)
        print(f"Total XP for {self.team_name}: {total_xp}")
        return

        
team_blue = Team("Team Blue")
team_x = Team("Team X")
team_blue.add_player("Nico")
team_blue.add_player("Lynn")
team_x.add_player("nick")
team_blue.show_players()
team_x.show_players()


# 팀에 없는 플레이어 제외 시도
team_blue.remove_player("Sam")


team_blue.show_XP()
team_x.show_XP()

```

> 첫번째와 마지막은 필요없는 정보를 가지고 있으므로 지워야 한다.

```python
from bs4 import BeautifulSoup
import cloudscraper

# Cloudscraper로 요청 보내기
scraper = cloudscraper.create_scraper()
url = "https://weworkremotely.com/remote-full-time-jobs#job-listings"
response = scraper.get(url)

# BeautifulSoup으로 HTML 파싱
soup = BeautifulSoup(response.content, "html.parser")

# "job_list"라는 id를 가진 div 태그 찾기
jobs= soup.find("section",class_="jobs").find_all("li")[1:-1]  # 그냥 find("li")라고 하면 첫번째것만 가져온다. 따라서 전부 가져오려면 find_all() 사용 여기에서는 제일 처음과 마지막은 필요없으므로 제외.

for job in jobs:
    title = job.find("span", class_ = "title").text   #title이라는 class를 가진 span을 요청 / .text > 텍스트만 요청해서 얻기! 안적으면 각종 정보가 전부 등장.
    # region = job.find("span", class_="region").text   #region이라는 class를 가진 span을 요청
    company, position, region = job.find_all("span", class_="company")  #find라고 하면 첫번째것만, find_all이라고 하면 list로 값들을 반환.
    # print(title,"---------",region)
    company = company.text
    position = position.text
    region = region.text
    print(title,'---',company,'---',position,'---',region,"---------\n")
```

```python
from bs4 import BeautifulSoup
import cloudscraper

class RemoteJobScraper:
    def __init__(self):
        self.scraper = cloudscraper.create_scraper()
        self.all_jobs = []

    def scrape_page(self, url):
        print(f"Scrapping {url}...")
        response = self.scraper.get(url)

        # 응답이 성공적이지 않을 경우 처리
        if response.status_code != 200:
            print(f"Failed to fetch page: {url}, Status code: {response.status_code}")
            return

        # BeautifulSoup으로 HTML 파싱
        soup = BeautifulSoup(response.content, "html.parser")

        # 'jobsboard' 테이블 찾기
        jobs_table = soup.find("table", id="jobsboard")
        if not jobs_table:
            print("Jobsboard table not found.")
            return

        # 'job' 클래스의 모든 행 찾기
        jobs = jobs_table.find_all("tr", class_="job")

        # 각 직업 정보 추출
        for job in jobs:
            # 제목, 회사 이름, 위치 및 URL 추출
            title_tag = job.find("h2", itemprop="title")
            name_tag = job.find("span", class_="companyLink")
            location_tag = job.find("div", class_="location")
            url_tag = job.find("a", class_="preventLink")

            # 각 태그가 존재하는지 확인하고, 존재할 경우 데이터 수집
            if title_tag and name_tag and location_tag and url_tag:
                title = title_tag.text.strip()
                company = name_tag.text.strip()
                location = location_tag.text.strip()
                job_url = f'https://remoteok.com{url_tag["href"]}'

                job_data = {
                    "title": title,
                    "company": company,
                    "location": location,
                    "url": job_url
                }
                self.all_jobs.append(job_data)
            else:
                print("Some job postings are missing title, company, location, or URL information.")

    def scrape_keywords(self, keywords):
        # 각 키워드에 대해 페이지 스크래핑
        for keyword in keywords:
            url = f"https://remoteok.com/remote-{keyword}-jobs"
            self.scrape_page(url)

    def get_all_jobs(self):
        # 수집된 모든 직업 정보 반환
        return self.all_jobs


# 클래스 사용 예제
if __name__ == "__main__":
    # 스크래퍼 인스턴스 생성
    job_scraper = RemoteJobScraper()

    # 키워드별로 스크래핑 호출
    keywords = ["flutter", "python", "golang"]
    job_scraper.scrape_keywords(keywords)

    # 수집된 모든 직업 정보 출력
    all_jobs_remote = job_scraper.get_all_jobs()
    print(f"Total jobs found: {len(all_jobs_remote)}")
    for job in all_jobs_remote:
        print(job)
```

```python
from bs4 import BeautifulSoup
import cloudscraper

class JobScraper:
    def __init__(self):
        self.scraper = cloudscraper.create_scraper()
        self.all_jobs = {}

    def scrape_page(self, url, category):
        print(f"Scrapping {url}...")
        response = self.scraper.get(url)
        soup = BeautifulSoup(response.content, "html.parser")
        jobs = soup.find("div", id="content").find_all("div", class_='bjs-jlid__wrapper')

        job_list = []
        for job in jobs:
            title = job.find("h4", class_="bjs-jlid__h")
            company = job.find("a", class_="bjs-jlid__b")
            explanation = job.find("div", class_="bjs-jlid__description")
            url = job.find("h4", class_="bjs-jlid__h").find("a")["href"]

            job_data = {
                "title": title.text.strip() if title else "No title",
                "company": company.text.strip() if company else "No company",
                "explanation": explanation.text.strip() if explanation else "No description",
                "url": f'https://berlinstartupjobs.com{url}' if url else "No URL"
            }
            job_list.append(job_data)

        if category in self.all_jobs:
            self.all_jobs[category].extend(job_list)
        else:
            self.all_jobs[category] = job_list

    def scrape_engineering_pages(self, base_url, pages, category):
        for i in range(1, pages + 1):
            page_url = f"{base_url}page/{i}/" if i > 1 else base_url
            self.scrape_page(page_url, category)

    def display_results(self):
        for category, jobs in self.all_jobs.items():
            print(f"\n=== {category.upper()} JOBS ===")
            for idx, job in enumerate(jobs, 1):
                print(f"\nJob {idx}:")
                print(f"Title: {job['title']}")
                print(f"Company: {job['company']}")
                print(f"Description: {job['explanation']}")
                print(f"URL: {job['url']}")
            print(f"\nTotal {category} jobs: {len(jobs)}")


scraper = JobScraper()
scraper.scrape_engineering_pages("https://berlinstartupjobs.com/engineering/", 2, "engineering")

keywords = {
    "python": "https://berlinstartupjobs.com/skill-areas/python/",
    "typescript": "https://berlinstartupjobs.com/skill-areas/typescript/",
    "javascript": "https://berlinstartupjobs.com/skill-areas/javascript/"
}

for category, url in keywords.items():
    scraper.scrape_page(url, category)

scraper.display_results()
```
