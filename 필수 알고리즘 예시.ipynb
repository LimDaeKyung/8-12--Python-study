#### 카데인 알고리즘 : 1차원 배열에서 연속된 부분 배열의 합 중 가장 큰 값을 찾는 알고리즘

```python
def max_sum(data):
    # 최대 부분합 계산을 위한 초기값 설정
    max_ending_here = max_so_far = data[0]  # 배열의 첫 번째 값을 기준으로 초기화

    # 배열의 두 번째 값부터 끝까지 순차적으로 확인
    for i in range(1, len(data)):
        max_ending_here = max(data[i], max_ending_here + data[i])
        max_so_far = max(max_so_far, max_ending_here)

    return max_so_far
```

```python
data = [-2, 1, -3, 4, -1, 2, 1, -5, 4]
result = max_sum(data)
print(f"최대 부분합: {result}")

```

```python
import random

# 상대 리스트 초기화
opponents = ["상대 1", "상대 2", "상대 3", "상대 4", "상대 5"]

# 가위 바위 보 결과 판별 함수
def determine_winner(user_choice, computer_choice):
    if user_choice == computer_choice:
        return "tie"  # 비김
    elif (user_choice == "가위" and computer_choice == "보") or \
         (user_choice == "바위" and computer_choice == "가위") or \
         (user_choice == "보" and computer_choice == "바위"):
        return "win"  # 사용자가 이김
    else:
        return "lose"  # 사용자가 짐

# 컴퓨터의 가위 바위 보 랜덤 선택
def computer_choice():
    return random.choice(["가위", "바위", "보"])

# 가위 바위 보 게임 시작
def start_game():
    current_opponent_idx = 0  # 첫 번째 상대부터 시작
    rounds = 0  # 총 대결 횟수
    battle_history = {opponent: 0 for opponent in opponents}  # 대결 기록 (상대 이름: 대결 횟수)

    while current_opponent_idx < len(opponents):
        current_opponent = opponents[current_opponent_idx]
        user_choice = input(f"당신의 선택 (가위, 바위, 보): ")
        comp_choice = computer_choice()
        rounds += 1
        
        print(f"{current_opponent}의 선택: {comp_choice}")
        
        result = determine_winner(user_choice, comp_choice)

        battle_history[current_opponent] += 1

        if result == "win":
            print(f"당신이 이겼습니다! {current_opponent}을(를) 이겼습니다.")
            if current_opponent_idx == len(opponents) - 1:  # 마지막 상대를 이겼을 때
                print("최종 상대를 이겼습니다! 게임 종료!")
                break
            current_opponent_idx += 1  # 다음 상대로 이동
        elif result == "lose":
            print(f"당신이 졌습니다. {current_opponent}에게 패배하여 이전 상대와 다시 대결합니다.")
            if current_opponent_idx > 0:  # 첫 번째 상대가 아니면 이전 상대로 돌아감
                current_opponent_idx -= 1
        else:
            print(f"비겼습니다. {current_opponent}와 다시 대결합니다.")
    
    # 최종 결과 출력
    print(f"\n총 대결 횟수: {rounds}")
    print("대결 기록:")
    for opponent, count in battle_history.items():
        print(f"{opponent}: {count}번 대결")

# 게임 실행
if __name__ == "__main__":
    start_game()

```

### 스택과 큐

- 입력과 출력을 한 방향으로 제한한 알고리즘
- 스택은 바닥부터 데이터를 차곡차곡 쌓는 개념
1. 푸시 : 그릇을 닦아서 찬장에 쌓아 놓는다.
2. 팝 : 그릇을 사용하기 위해 찬장에서 가장 위에 있는 그릇을 꺼내 온다.
> LIFO(last in first out) :마지막에 들어간 것이 제일 처음 나온다.

```python
# 스택 구현
def push(item):
    stack.append(item)

def pop():
    return stack.pop()

if __name__ =='__main__':
    stack = []
    push(1)
    push(2)
    push(3)
    push(4)

    print('현재 stack의 모습')
    print(stack)

    while stack:
        print('pop > {}'.format(pop()))
```

> 스택의 응용 : 계산기

#### 큐
> FIFO(first in first out): 처음 저장한 데이터를 먼저 사용하는 방식.
1. put : 매표소에 줄을 선다
2. out : 줄을 선 순서대로 앞에서부터 표를 산다.
> 스택과 달리 배열을 사용하는 것이 조금 더 편리하다.

```python
# 큐 구현
def push(item):
    stack.append(item)

def get():
    return stack.pop(0)

if __name__ =='__main__':
    stack = []
    push(1)
    push(2)
    push(3)
    push(4)

    print('현재 stack의 모습')
    print(stack)

    while stack:
        print('pop > {}'.format(get()))
```

#### 트리 구조
> 트리구조는 다른 자료 구조보다 자료를 저장하거나 검색하는 등의 방법이 간단하며 메모리를 효율적으로 사용하기 때문이다.

1. 최상위 노드 : 루트 노드 
2. 최하위 노드 : 리프 노드 
3. 자신의 노드보다 상위에 있는 노드 : 부모 노드 
4. 자신의 노드 보다 하위에 있는 노드 : 자식 노드
5. 같은 부모노드를 갖는 노드들의 사이를 형제 노드

> 트리 구조는 레벨과 높이가 존재한다.
- 자식노드는 몇개라도 상관 없지만 부모 노드는 반드시 하나만 존재해야 한다. 
- 레벨은 루트 노드부터 해당 노드까지 경로를 찾아 오는데 방문한 총 노드의 수가 된다. 즉 최상위부터 0에서 시작하여 1씩 증가하면서 커지는 것

> 트리의 용어
(1)차수 : 한 노드에 연결된 서브 트리의 개수를 차수라고 한다. 이때 트리의 차수가 2 이하인 트리를 이진 트리라고 한다.

> 이진트리 (Binary Tree)
- 자식 노드를 두개 이하만 갖는 트리. 즉 차수가 2 이하인 트리.
